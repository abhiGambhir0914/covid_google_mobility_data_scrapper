```python
i=0
j=0
domains = []

for html_file in glob.glob("scraper/files/full/*.html"):
#     print(html_file)    
    f =  open(html_file)
    sel = Selector(text=f.read())
    
    print(html_file)    
    
    # common in all files
    percentages = sel.xpath('//span[contains(@style, "font-family: GoogleSans-Regular; font-size:44px")]/text()').extract()
#      or contains(@style, "font-family: GoogleSans-Medium; font-size:13px")
   
    
    country_name = sel.xpath('//span[contains(@style, "font-family: GoogleSans-Regular; font-size:22px")]/text()')[0].extract()
    country_name = country_name.split('  ')[0]
    print(country_name)
    
    
    all_provinces_values = sel.xpath('//span[contains(@style, "font-family: GoogleSans-Medium; font-size:13px") or contains(@style, "font-family: GoogleSans-Medium; font-size:11px")]/text()').extract()
    all_provinces_values = list(filter(lambda a: a != 'Not enough data for this date', all_provinces_values))
    n = 6
    
    
    percentages = list(map(lambda i: i.strip(), percentages))
    

    if i == 0:
        domains = sel.xpath('//span[contains(@style, "font-family: GoogleSans-Regular; font-size:44px")]/preceding::div[1]/span/text()').extract()
        domains = list(map(lambda i: i.strip(), domains))
        
        df1 = pd.DataFrame([percentages],columns=domains, index=[country_name])
        i=1
        j=1
    else:
        df2 = pd.DataFrame([percentages],columns=domains, index=[country_name])
        df1 = pd.concat([df1, df2], axis=0)
        
        
# files containing provinces/states

    if all_provinces_values:
        provinces_name = sel.xpath('//span[contains(@style, "font-family: GoogleSans-Regular; font-size:20px")]/text()').extract()
        provinces_name = list(map(lambda i: i.strip(), provinces_name))
        
#         print(provinces_name)
        divide_final_province = [all_provinces_values[i * n:(i + 1) * n] for i in range((len(all_provinces_values) + n - 1) // n )]  
    
        print()
        print(divide_final_province)
        print()
        
        for province in divide_final_province:
#             print(province)
            if j == 0:
                df1 = pd.DataFrame([province],columns=domains, index=[provinces_name[divide_final_province.index(province)]])
                j=1
                i=1
            else:
                df2 = pd.DataFrame([province],columns=domains, index=[provinces_name[divide_final_province.index(province)]])
                df1 = pd.concat([df1, df2], axis=0)
    else: 
        print('**'*20,'no province')
        print()
        
print()        
# display(df1)
```
